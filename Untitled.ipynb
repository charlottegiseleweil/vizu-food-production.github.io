```python
graph_path = 'data/graph/'

country_ssp1 = pd.read_csv(graph_path + 'graph_country_data_ssp1.csv', na_values=[0, np.inf])
country_ssp2 = pd.read_csv(graph_path + 'graph_country_data_ssp2.csv', na_values=[0, np.inf])
country_ssp3 = pd.read_csv(graph_path + 'graph_country_data_ssp3.csv', na_values=[0, np.inf])
country_ssp4 = pd.read_csv(graph_path + 'graph_country_data_ssp4.csv', na_values=[0, np.inf])
country_ssp5 = pd.read_csv(graph_path + 'graph_country_data_ssp5.csv', na_values=[0, np.inf])

continent_ssp1 = pd.read_csv(graph_path + 'graph_continent_data_ssp1.csv', na_values=[0, np.inf])
continent_ssp2 = pd.read_csv(graph_path + 'graph_continent_data_ssp2.csv', na_values=[0, np.inf])
continent_ssp3 = pd.read_csv(graph_path + 'graph_continent_data_ssp3.csv', na_values=[0, np.inf])
continent_ssp4 = pd.read_csv(graph_path + 'graph_continent_data_ssp4.csv', na_values=[0, np.inf])
continent_ssp5 = pd.read_csv(graph_path + 'graph_continent_data_ssp5.csv', na_values=[0, np.inf])

country_ssp1.head()
```

```python
columns = ['Country', 'Calories2000', 'Population2000', 'Caloryrequirements2000',
       'Cropland2000', 'Sufficiency2000', 'Percentageoftotalcal2000',
       'Yield2000', 'Calories2050', 'Population2050', 'Caloryrequirements2050',
       'Cropland2050', 'Sufficiency2050', 'Percentageoftotalcal2050',
       'Yield2050', 'diffCalories', 'diffPopulation', 'diffCaloryrequirements',
       'diffYield']
```

```python
country_ssp1.columns = columns
country_ssp2.columns = columns
country_ssp3.columns = columns
country_ssp4.columns = columns
country_ssp5.columns = columns

continent_ssp1.columns = columns
continent_ssp2.columns = columns
continent_ssp3.columns = columns
continent_ssp4.columns = columns
continent_ssp5.columns = columns


world_ssp1 = pd.DataFrame(country_ssp1.apply({'Country':np.sum, 'Calories2000':np.sum, 'Population2000':np.sum, 'Caloryrequirements2000':np.sum,
       'Cropland2000':np.mean, 'Sufficiency2000':np.mean, 'Percentageoftotalcal2000':np.mean,
       'Yield2000':np.sum, 'Calories2050':np.sum, 'Population2050':np.sum, 'Caloryrequirements2050':np.sum,
       'Cropland2050':np.mean, 'Sufficiency2050':np.mean, 'Percentageoftotalcal2050':np.mean,
       'Yield2050':np.mean, 'diffCalories':np.mean, 'diffPopulation':np.mean, 'diffCaloryrequirements':np.mean,
       'diffYield':np.mean})).transpose()
world_ssp2 = pd.DataFrame(country_ssp2.apply({'Country':np.sum, 'Calories2000':np.sum, 'Population2000':np.sum, 'Caloryrequirements2000':np.sum,
       'Cropland2000':np.mean, 'Sufficiency2000':np.mean, 'Percentageoftotalcal2000':np.mean,
       'Yield2000':np.sum, 'Calories2050':np.sum, 'Population2050':np.sum, 'Caloryrequirements2050':np.sum,
       'Cropland2050':np.mean, 'Sufficiency2050':np.mean, 'Percentageoftotalcal2050':np.mean,
       'Yield2050':np.mean, 'diffCalories':np.mean, 'diffPopulation':np.mean, 'diffCaloryrequirements':np.mean,
       'diffYield':np.mean})).transpose()
world_ssp3 = pd.DataFrame(country_ssp3.apply({'Country':np.sum, 'Calories2000':np.sum, 'Population2000':np.sum, 'Caloryrequirements2000':np.sum,
       'Cropland2000':np.mean, 'Sufficiency2000':np.mean, 'Percentageoftotalcal2000':np.mean,
       'Yield2000':np.sum, 'Calories2050':np.sum, 'Population2050':np.sum, 'Caloryrequirements2050':np.sum,
       'Cropland2050':np.mean, 'Sufficiency2050':np.mean, 'Percentageoftotalcal2050':np.mean,
       'Yield2050':np.mean, 'diffCalories':np.mean, 'diffPopulation':np.mean, 'diffCaloryrequirements':np.mean,
       'diffYield':np.mean})).transpose()
world_ssp4 = pd.DataFrame(country_ssp4.apply({'Country':np.sum, 'Calories2000':np.sum, 'Population2000':np.sum, 'Caloryrequirements2000':np.sum,
       'Cropland2000':np.mean, 'Sufficiency2000':np.mean, 'Percentageoftotalcal2000':np.mean,
       'Yield2000':np.sum, 'Calories2050':np.sum, 'Population2050':np.sum, 'Caloryrequirements2050':np.sum,
       'Cropland2050':np.mean, 'Sufficiency2050':np.mean, 'Percentageoftotalcal2050':np.mean,
       'Yield2050':np.mean, 'diffCalories':np.mean, 'diffPopulation':np.mean, 'diffCaloryrequirements':np.mean,
       'diffYield':np.mean})).transpose()
world_ssp5 = pd.DataFrame(country_ssp5.apply({'Country':np.sum, 'Calories2000':np.sum, 'Population2000':np.sum, 'Caloryrequirements2000':np.sum,
       'Cropland2000':np.mean, 'Sufficiency2000':np.mean, 'Percentageoftotalcal2000':np.mean,
       'Yield2000':np.sum, 'Calories2050':np.sum, 'Population2050':np.sum, 'Caloryrequirements2050':np.sum,
       'Cropland2050':np.mean, 'Sufficiency2050':np.mean, 'Percentageoftotalcal2050':np.mean,
       'Yield2050':np.mean, 'diffCalories':np.mean, 'diffPopulation':np.mean, 'diffCaloryrequirements':np.mean,
       'diffYield':np.mean})).transpose()


worlds = [world_ssp1, world_ssp2, world_ssp3, world_ssp4, world_ssp5]

for world in worlds:
    world['Country'] = 'World'
    world['diffCalories'] = 100*(world['Calories2050'] - world['Calories2000']) / world['Calories2000']
    world['diffPopulation'] = 100*(world['Population2050'] - world['Population2000']) / world['Population2000']
    world['diffCaloryrequirements'] = 100*(world['Caloryrequirements2050'] - world['Caloryrequirements2000']) / world['Caloryrequirements2000']
```

```python
ssp1 = pd.concat([country_ssp1, continent_ssp1, world_ssp1]).reset_index(drop=True)
ssp2 = pd.concat([country_ssp2, continent_ssp2, world_ssp2]).reset_index(drop=True)
ssp3 = pd.concat([country_ssp3, continent_ssp3, world_ssp3]).reset_index(drop=True)
ssp4 = pd.concat([country_ssp4, continent_ssp4, world_ssp4]).reset_index(drop=True)
ssp5 = pd.concat([country_ssp5, continent_ssp5, world_ssp5]).reset_index(drop=True)
```

```python
data1 = pd.merge(ssp1, gdp, how='outer', on='Country')
data2 = pd.merge(ssp2, gdp, how='outer', on='Country')
data3 = pd.merge(ssp3, gdp, how='outer', on='Country')
data4 = pd.merge(ssp4, gdp, how='outer', on='Country')
data5 = pd.merge(ssp5, gdp, how='outer', on='Country')
```

```python
datas = [data1, data2, data3, data4, data5]
for data in datas:
    data['diffCropland'] = 100*(data['Cropland2050'] - data['Cropland2000']) / data['Cropland2000']
```

```python
data1.to_csv(graph_path + 'SSP1.csv')
data2.to_csv(graph_path + 'SSP2.csv')
data3.to_csv(graph_path + 'SSP3.csv')
data4.to_csv(graph_path + 'SSP4.csv')
data5.to_csv(graph_path + 'SSP5.csv')
```
