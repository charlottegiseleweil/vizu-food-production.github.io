```python
import pandas as pd
import numpy as np
import math

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('SSP1_cc_countries.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['total_calories'] = df['calories_per_ha 2000'] * df['ha_per_cell_5m']
```

```python
df['cal_diff'] = (df['calories_per_ha'] - df['calories_per_ha 2000']) * df['ha_per_cell_5m']
```

```python
bibi = df.groupby('country')['calories_per_ha'].sum()
bibi.loc['Qatar']
```

```python
bibi2 = df.groupby('country')['calories_per_ha 2000'].sum()
bibi2.loc['Qatar']
```

```python
df2 = df.groupby('country')[['cal_diff', 'total_calories']].sum()
df2.loc['Qatar', :]
```

```python
pd.options.display.float_format = '{:.2f}'.format
ranking = (100 * df2['cal_diff'] / df2['total_calories']).round(2).sort_values()
ranking
```
