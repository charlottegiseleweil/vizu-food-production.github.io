```python
import pandas as pd
import numpy as np
import math


import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path = 'data/originals/'
country_path = 'data/countries/'
graph_path = 'data/graph/'
with_country_path = 'data/with_country/'
# climate_models = ['gs', 'he', 'mr', 'cc']
climate_model = 'he'
```

```python
cal_per_year_person = 2.355e6 * 365
```

```python
data_2000 = pd.read_csv(with_country_path + '2000_country.csv', index_col='pixel_id')
data_ssp1 = pd.read_csv(with_country_path + 'SSP1_country.csv', index_col='pixel_id')
data_ssp2 = pd.read_csv(with_country_path + 'SSP2_country.csv', index_col='pixel_id')
data_ssp3 = pd.read_csv(with_country_path + 'SSP3_country.csv', index_col='pixel_id')
data_ssp4 = pd.read_csv(with_country_path + 'SSP4_country.csv', index_col='pixel_id')
data_ssp5 = pd.read_csv(with_country_path + 'SSP5_country.csv', index_col='pixel_id')

scenarios = [data_ssp1, data_ssp2, data_ssp3, data_ssp4, data_ssp5]
scenario_names = ['SSP1', 'SSP2', 'SSP3', 'SSP4', 'SSP5']
```

```python
# Country 2000
total_cal_2000 = data_2000['Calories 2000'].sum()
cal_country_2000 = data_2000.groupby('country')['Calories 2000'].sum()
cal_perc_country_2000 = 100 * cal_country_2000 / total_cal_2000
population_country_2000 = data_2000.groupby('country')['Population 2000'].sum()
cropland_per_country_2000 = 100 * data_2000.groupby('country')['%cropland 2000'].mean()
cal_requirements_country_2000 = population_country_2000 * cal_per_year_person
sufficiency_country_2000 = 100 * cal_country_2000 / cal_requirements_country_2000
cropland_yield_country_2000 = data_2000.groupby('country')['cal_per_ha 2000'].mean()
country_2000 = pd.concat([cal_country_2000, population_country_2000, cal_requirements_country_2000, cropland_per_country_2000, sufficiency_country_2000, cal_perc_country_2000, cropland_yield_country_2000], axis=1) 
country_2000.columns = ['Calories2000', 'Population2000', 'Calory_requirements_2000', 'Cropland2000', 'Sufficiency2000', 'Percentage_of_total_cal2000', 'Yield2000']
country_2000.head()
```

```python
# Continent 2000
cal_continent_2000 = data_2000.groupby('continent')['Calories 2000'].sum()
cal_perc_continent_2000 = 100 * cal_continent_2000 / total_cal_2000
population_continent_2000 = data_2000.groupby('continent')['Population 2000'].sum()
cropland_per_continent_2000 = 100 * data_2000.groupby('continent')['%cropland 2000'].mean()
cal_requirements_continent_2000 = population_continent_2000 * cal_per_year_person
sufficiency_continent_2000 = 100 * cal_continent_2000 / cal_requirements_continent_2000
cropland_yield_continent_2000 = data_2000.groupby('continent')['cal_per_ha 2000'].mean()
continent_2000 = pd.concat([cal_continent_2000, population_continent_2000, cal_requirements_continent_2000, cropland_per_continent_2000, sufficiency_continent_2000, cal_perc_continent_2000, cropland_yield_continent_2000], axis=1) 
continent_2000.columns = ['Calories2000', 'Population2000', 'Calory_requirements_2000', 'Cropland2000', 'Sufficiency2000', 'Percentage_of_total_cal2000', 'Yield2000']
continent_2000
```

```python
demand_2000 = country_2000['Calory requirements 2000'].sum()
production_2000 = country_2000['Calories 2000'].sum()
print('World demand 2000: {:,}\nWorld production 2000: {:,}'.format(demand_2000, production_2000))
```

```python
for scenario, name in zip(scenarios, scenario_names):
    # Country 2050
    total_cal = scenario['Calories 2050'].sum()
    cal_country = scenario.groupby('country')['Calories 2050'].sum()
    cal_perc_country = 100 * cal_country / total_cal
    population_country = scenario.groupby('country')['Population 2050'].sum()
    cropland_per_country = 100 * scenario.groupby('country')['%cropland 2050'].mean()
    cal_requirements_country = population_country * cal_per_year_person
    sufficiency_country = 100 * cal_country / cal_requirements_country
    cropland_yield_country = scenario.groupby('country')['cal_per_ha 2050'].mean()
    country = pd.concat([cal_country, population_country, cal_requirements_country, cropland_per_country, sufficiency_country, cal_perc_country, cropland_yield_country], axis=1) 
    country.columns = ['Calories 2050', 'Population 2050', 'Calory requirements 2050', 'Cropland (%) 2050', 'Sufficiency (%) 2050', 'Percentage of total cal 2050', 'Yield 2050']
    
    # Delta country 2000 - 2050
    delta_cal_country = 100*scenario.groupby('country')['ΔCalories'].sum() / country_2000['Calories 2000']
    delta_pop_country = 100*(country['Population 2050'] - country_2000['Population 2000'])/country_2000['Population 2000']
    delta_yield_country = 100*(country['Yield 2050'] - country_2000['Yield 2000']) / country_2000['Yield 2000']
    delta_cal_req_country = 100*(country['Calory requirements 2050'] - country_2000['Calory requirements 2000']) / country_2000['Calory requirements 2000']
    delta_country = pd.concat([delta_cal_country, delta_pop_country, delta_cal_req_country, delta_yield_country], axis=1)
    delta_country.columns = ['ΔCalories (%)', 'ΔPopulation (%)', 'ΔCalory requirements (%)', 'ΔYield (%)']
    country = country_2000.join(country, on='country').join(delta_country, on='country')
    
    
    # Continent 2050
    cal_continent = scenario.groupby('continent')['Calories 2050'].sum()
    cal_perc_continent = 100 * cal_continent / total_cal
    population_continent = scenario.groupby('continent')['Population 2050'].sum()
    cropland_per_continent = 100 * scenario.groupby('continent')['%cropland 2050'].mean()
    cal_requirements_continent = population_continent * cal_per_year_person
    sufficiency_continent = 100 * cal_continent / cal_requirements_continent
    cropland_yield_continent = scenario.groupby('continent')['cal_per_ha 2050'].mean()
    continent = pd.concat([cal_continent, population_continent, cal_requirements_continent, cropland_per_continent, sufficiency_continent, cal_perc_continent, cropland_yield_continent], axis=1) 
    continent.columns = ['Calories 2050', 'Population 2050', 'Calory requirements 2050', 'Cropland (%) 2050', 'Sufficiency (%) 2050', 'Percentage of total cal 2050', 'Yield 2050']
    
    # Delta continent 2050
    delta_cal_continent = 100*scenario.groupby('continent')['ΔCalories'].sum() / continent_2000['Calories 2000']
    delta_pop_continent = 100*(continent['Population 2050'] - continent_2000['Population 2000']) / continent_2000['Population 2000']
    delta_yield_continent = 100*(continent['Yield 2050'] - continent_2000['Yield 2000']) / continent_2000['Yield 2000']
    delta_cal_req_continent = 100*(continent['Calory requirements 2050'] - continent_2000['Calory requirements 2000']) / continent_2000['Calory requirements 2000']
    delta_continent = pd.concat([delta_cal_continent, delta_pop_continent, delta_cal_req_continent, delta_yield_continent], axis=1)
    delta_continent.columns = ['ΔCalories (%)', 'ΔPopulation (%)', 'ΔCalory requirements (%)', 'ΔYield (%)']
    continent = continent_2000.join(continent, on='continent').join(delta_continent, on='continent')
     
    country.to_csv(graph_path + 'graph_country_' + name + '.csv')
    continent.to_csv(graph_path + 'graph_continent_' + name + '.csv')
    
    demand_2050 = country['Calory requirements 2050'].sum()
    production_2050 = country['Calories 2050'].sum()
    print('{}:\nWorld demand 2050: {:,}\nWorld production 2050: {:,}\n'.format(name, demand_2050, production_2050))
```
